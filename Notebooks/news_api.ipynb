```python
import requests
from bs4 import BeautifulSoup

# url = news_url_access[3]
def fallback_scrape(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Try common article tags
        paragraphs = soup.find_all('p')
        full_text = ' '.join([p.get_text() for p in paragraphs])
        return full_text.strip()
    except Exception as e:
        print(f"[!] Fallback scrape failed: {e}")
        return ""
# t = fallback_scrape(url)
```

```python
# for i in df['newsUrl']:
df['summary'] = df['newsUrl'].apply(lambda x: fallback_scrape(x))
# print(news_url_access)
```

```python
def safe_fallback_scrape(url):
	try:
		resp = requests.get(url)
		if resp.status_code == 200:
			return fallback_scrape(url)
		else:
			return " "
	except Exception as e:
		print(f"[!] Error fetching {url}: {e}")
		return " "

df['summary'] = df['newsUrl'].map(safe_fallback_scrape)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

data_list = []

for i in links:
    soup = BeautifulSoup(requests.get(i).text, 'html.parser')
    title = next((h1.get_text(strip=True) for h1 in soup.find_all('h1', class_='title')), "")
    subtitle = next((h2.get_text(strip=True) for h2 in soup.find_all('h2', class_='sub-title')), "")
    image = next((img.find('source')['srcset'] for img in soup.find_all('div', class_='picture verticle') if img.find('source')), "")
    updated_time = next((t.get_text(strip=True) for t in soup.find_all('p', class_='updated-time')), "")
    publisher = next((a.get_text(strip=True)+', The Hindu' for a in soup.find_all('a', class_='person-name no-lnk')), "")

    data_list.append({
        'title': title,
        'subtitle': subtitle,
        'image': image,
        'updated_time': updated_time,
        'author': publisher,
        'url': i
    })

df_articles = pd.DataFrame(data_list)
print(df_articles.head())
```

```python
import feedparser

# RSS Feeds from top tech news sources
rss_feeds = {
    "The Verge": "https://www.theverge.com/rss/index.xml",
    "TechCrunch": "https://techcrunch.com/feed/",
    "BBC Technology": "http://feeds.bbci.co.uk/news/technology/rss.xml",
}

def fetch_tech_articles():
    articles = []
    for source, url in rss_feeds.items():
        feed = feedparser.parse(url)
        for entry in feed.entries[:3]:  # top 3 articles per source
            articles.append({
                "source": source,
                "title": entry.title,
                "link": entry.link,
                "published": entry.published,
                "summary": entry.summary
            })
    return articles

# Test: print latest 3 articles
if __name__ == "__main__":
    news = fetch_tech_articles()
    for n in news:
        print(f"\n📌 {n['title']} ({n['source']})")
        print(f"Published: {n['published']}")
        print(f"Summary: {n['summary']}")
        print(f"Link: {n['link']}")
```

```python
import feedparser
import pandas as pd
import re

# Tech RSS feeds
rss_feeds = {
    "TechCrunch": "https://techcrunch.com/feed/",
    "The Verge": "https://www.theverge.com/rss/index.xml",
    "BBC Technology": "http://feeds.bbci.co.uk/news/technology/rss.xml"
}

def get_image_url(entry):
    """Try to extract image URL from media tags or summary HTML."""
    if "media_content" in entry:
        return entry.media_content[0].get("url")
    elif "media_thumbnail" in entry:
        return entry.media_thumbnail[0].get("url")
    elif "summary" in entry and "<img" in entry.summary:
        match = re.search(r'<img.*?src="(.*?)"', entry.summary)
        if match:
            return match.group(1)
    return None

def fetch_articles_df():
    all_articles = []

    for source, url in rss_feeds.items():
        feed = feedparser.parse(url)

        for entry in feed.entries[:5]:  # Top 5 articles per source
            article = {
                "source": source,
                "title": entry.get("title", "N/A"),
                "summary": re.sub("<[^<]+?>", "", entry.get("summary", ""))[:300],  # remove HTML
                "published": entry.get("published", "N/A"),
                "link": entry.get("link", "N/A"),
                "image": get_image_url(entry)
            }
            all_articles.append(article)

    # Convert to DataFrame
    df = pd.DataFrame(all_articles)
    return df

# 🔍 Example usage
if __name__ == "__main__":
    df = fetch_articles_df()
    print(df.head())
    # Optional: Save to CSV
    df.to_csv("tech_news_feed.csv", index=False)
```
